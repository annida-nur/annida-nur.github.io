# **Packages, Modules & Libraries**

```python
import pandas as pd
import folium
import base64
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.ticker as ticker
from folium import IFrame
```

# Data

```python
region = pd.read_excel('data_uts.xlsx')
print(region.shape)
region
```

```python
data = pd.read_excel('data_uts(2).xlsx')
print(data.shape)
data
```

# Area of Interest

```python
m = folium.Map(
    zoom_start=18,
    location=[-7.433524029013859, 109.25046078786383]
)
m
```

# Data Visualization & Marking Area

```python
resolution, width, height = 75, 7, 3
station = '42'
```

**Petak 1**

```python
nama = region.iloc[0]['nomor']
warna = region.iloc[0]['color']
lat = region.iloc[0]['latitude']
lon = region.iloc[0]['longitude']

sns.set_style('darkgrid')
sns.set(rc={'figure.figsize':(8,5)})

ax = sns.lineplot(data=data, x ='Month', y = 'Koordinat 1')

ax.xaxis.set_major_locator(ticker.MultipleLocator(1))
plt.title('Time Series NDVI Koordinat 1')
plt.ylabel('NDVI Value')
plt.xlabel('Month')
png = 'mpld3_1.png'.format(station)
plt.savefig(png, dpi=resolution)

encoded = base64.b64encode(open(png, 'rb').read())

html = '<img src="data:image/png;base64,{}">'.format
iframe = IFrame(html(encoded.decode('UTF-8')), width=(width*resolution)+100, height=(height*resolution)+200)
popup = folium.Popup(iframe, max_width=5000)

icon = folium.Icon(color=warna, icon="ok")
marker = folium.Marker(
    location=[lat, lon], 
    tooltip=(f"Lokasi : {nama} || Latitude : {lat} || Longitude : {lon}"),
    popup=popup, 
    icon=icon)
marker.add_to(m)

m
```
